```python
date = pd.read_csv(r'air_date-utf8.csv')
```

```python
date.drop_duplicates(inplace= Ture)
filter_date = date[['FFP_DATE','LOAD_TIME','FLIGHT_COUNT','SUM_YR_1','SUM_YR_2','SEG_KM_SUM','AVG_INTERVAL','MAX_INTERVAL','avg_diiscount']]
filter_date['SUM_YR_1'].fillna(filter_date['SUM_YR_1'].mean(),inplace=Ture)
filter_date['SUM_YR_2'].fillna(filter_date['SUM_YR_2'].mean(),inplace=Ture)
```

```python
filter_date['入会时间'] = (pd.to_datetime(filter_date['LOAD_TIME'])-pd.to_datetime(filter_date['FFP_DATE'])).dt.days
filter_date['飞行次数']= filter_date['FLIGHT_COUNT']
filter_date['总里程']= filter_date['SEG_KM_SUM']
filter_date['平均每公里票价']= (filter_date['SUM_YR_1']+filter_date['SUM_YR_2'])/filter_date['SEG_KM_SUM']
filter_date['时间间隔差值']= filter_date['MAX_INTERVAL']-filter_date['AVG_INTERVAL']
filter_date['平均折扣率']= filter_date['avg_discount']
X = filter_date.iloc[:,-6:]
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler().fit(X)
```

```python
X_std = pd.DateFrame(std.transform(X),columns=X.columns)
```

```python
import sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
K = KMeans()
K.fit(X_std)
```

```python
K.labels_
sil=[]
for i in range(2,6):
    K_ = KMeans(i).fit(X_std)
    sil.append(silhouette_score(X_std,K_.labels_))
```
